`np.random.normal` makes possible drawing random samples from a normal (Gaussian) distribution. 

```python
np.random.normal(loc=10, scale=1)
```

We've drawn a random sample from a normal distribution in which mean is 10 and standard deviation is 1. Let's draw 10 more samples.

```python
for i in range(10):
    drawn_sample = np.random.normal(10, 1)
    print(drawn_sample)
```

Is there a better way when we drawn samples more than 1? Yes!

```python
np.random.normal(loc=10, scale=1, size=10)
```
